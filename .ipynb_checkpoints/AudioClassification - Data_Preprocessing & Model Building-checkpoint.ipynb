## Extract Features

Here we will be using Mel-Frequency Cepstral Coefficients(MFCC) from the audio samples. The MFCC summarises the frequency distribution across the window size, so it is possible to analyse both the frequency and time characteristics of the sound. These audio representations will allow us to identify features for classification.

```python
# Extracting MFCC's for every audio file

import pandas as pd
import os
import librosa

audio_dataset_path = 'UrbanSound8K/audio/'
metadata = pd.read_csv("UrbanSound8K/metadata/UrbanSound8K.csv")

metadata.head()
```

```python
def feature_extracter(file_name) :
    audio, sample_rate = librosa.load(file_name ,res_type = 'kaiser_fast')
    mfccs_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
    mfccs_scaled_features = np.mean(mfccs_features.T, axis=0)     # t -> means Transpose
    
    return mfccs_scaled_features
```

```python
import numpy as np
from tqdm import tqdm

extracted_features = []

for index_num,row in tqdm(metadata.iterrows()) :
    file_name = os.path.join(os.path.abspath(audio_dataset_path),'fold'+str(row["fold"])+'/',str(row["slice_file_name"]))
    final_class_labels = row["class"]
    data = feature_extracter(file_name)
    extracted_features.append([data, final_class_labels])
```

```python
### converting extracted_features to Pandas dataframe
extracted_features_df=pd.DataFrame(extracted_features,columns=['feature','class'])
extracted_features_df.head()
```

```python
### Split the dataset into independent and dependent dataset
X=np.array(extracted_features_df['feature'].tolist() )
y=np.array(extracted_features_df['class'].tolist() )
```

```python
X.shape
```

```python
y
```

```python
#label encoding

from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
y = to_categorical(labelencoder.fit_transform(y))
```

```python
y
```

```python
y.shape
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

# Model Creation

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from sklearn import metrics
```

```python
#number of classes 
num_labels = y.shape[1]
```

```python
num_labels
```

```python
model = Sequential()

#first layer
model.add(Dense(100,input_shape= (40,)))
model.add(Activation("relu"))
model.add(Dropout(0.3))

#second layer
model.add(Dense(200))
model.add(Activation("relu"))
model.add(Dropout(0.3))

#third layer
model.add(Dense(100))
model.add(Activation("relu"))
model.add(Dropout(0.3))

#final layer
model.add(Dense(num_labels,activation = "softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer="adam")
```

```python
#training the model

from tensorflow.keras.callbacks import ModelCheckpoint
from datetime import datetime

num_epochs = 100
num_batch_size = 32

checkpointer = ModelCheckpoint(filepath='saved_models/audio_classification.hdf5', verbose=1, save_best_only=True)

start = datetime.now()

model.fit(X_train, y_train, batch_size=num_batch_size, epochs=num_epochs,
           validation_data=(X_test,y_test), callbacks=[checkpointer], verbose=1)

duration = datetime.now() - start
print("Training completed in time : ", duration)
```

```python
test_accuracy = model.evaluate(X_test,y_test, verbose=1)
```

### Testing Some test audio data

Steps :
    
    1) preprocess the new audio data
    
    2) predict the classes
    
    3) Inverse transform your predicted label

```python


filename="UrbanSound8K/dog_bark.wav"
audio, sample_rate = librosa.load(filename, res_type='kaiser_fast') 
mfccs_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
mfccs_scaled_features = np.mean(mfccs_features.T,axis=0)

print(mfccs_scaled_features)
mfccs_scaled_features=mfccs_scaled_features.reshape(1,-1)
print(mfccs_scaled_features)
print(mfccs_scaled_features.shape)
predicted_label=model.predict_classes(mfccs_scaled_features)
print(predicted_label)
prediction_class = labelencoder.inverse_transform(predicted_label) 
prediction_class

```
